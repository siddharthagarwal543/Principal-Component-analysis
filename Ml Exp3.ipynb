```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

#Importing datsets
dataset=pd.read_csv("exe1.csv")
x=dataset.iloc[:,0:2].values
x
```

```python
#feature scaling
ss=StandardScaler()
x1=ss.fit_transform(x)
x1
```

```python
#applying pca
pca=PCA(n_components=2)
x2=pca.fit_transform(x1)
x2
```

```python
#Finding principal components
explained_variance=pca.explained_variance_
explained_variance_ratio=pca.explained_variance_ratio_
print(explained_variance," ",explained_variance_ratio)
```

Observation: As per the variance values we can say the second-dimension varianceis too low or even negligible and hence the dimension reduced was for the first dimension and the second dimension could be ignored, reducing dimensionality complexity.

3)Repeat the steps of task #2 for the dataset from https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data Use first two principal components. Visualise the transformed data.

```python
#Importing datsets
dataset=pd.read_csv("iris.csv")
x=dataset.iloc[:,0:3].values
x
```

```python
#Feature Scaling
ss=StandardScaler()
x1=ss.fit_transform(x)
x1
```

```python
#applying PCA
pca=PCA(n_components=2)
x2=pca.fit_transform(x1)
x2
```

```python
#Finding principal components
explained_variance=pca.explained_variance_
explained_variance_ratio=pca.explained_variance_ratio_
print(explained_variance," ",explained_variance_ratio)
```

```python
#Data visulaiztion
plt.scatter(x[:,0],x[:,1])
plt.legend('Original data')
plt.scatter(x2[:,0],x2[:,1])
plt.legend('tranformed data')
plt.show()
```

Conslusion: As per the variance values and graphs, the second dimension used had a low variance compared to the first dimension, and hence the variance of the first dimension was majorly reduced as seen from graph.
